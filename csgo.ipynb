**sobre o cs**:Counter-Strike (também abreviado por CS) é um popular jogo eletrônico de tiro em primeira pessoa. O jogo é baseado em rodadas nas quais equipes de contraterroristas e terroristas combatem-se até a eliminação completa de um dos times, e tem como objetivo principal plantar e desarmar bombas, ou sequestrar e salvar reféns.

```python
import pandas as pd
```

```python
df_jogador.columns = ['nome', 'pais', 'times', 'total mapas', 'total rounds', 'kd diff', 'kd', 'rating' ]
```

# dicionário das colunas

coluna 'nome': Se refere ao nick do jogadores no jogo.

coluna 'pais': Se refere a nacionalidade do jogador.

coluna 'times': Se refere aos times que o jogador jogou.

coluna 'total mapas': Se refere ao total de mapas jogados

coluna 'total rounds': Se refere ao total de rounds jogados

coluna 'kd diff': Se refere diferença existente entre os abates e as mortes.

coluna 'kd': Se refere a taxa Kills/Deaths (K/D)

coluna 'rating': Se refere a uma estatística que ajuda medir a performance e resultados de jogadores

```python
df_jogador.head()
```

```python
df_jogador = df_jogador[df_jogador['pais'] == 'Brazil']
```

```python
df_jogador.sort_values(by='kd', ascending=False).head(10)
```

# Análisando os jogadores com mais mapas jogados

```python
df_top10_mapas = df_jogador.sort_values(by='total mapas', ascending=False).head(10)
```

```python
ax = df_top10_mapas.plot(x = 'nome', 
                         y = 'total mapas', 
                         kind ='bar',
                         color='orange', 
                         fontsize=15,
                         figsize =(18, 6))
ax.set_title('jogadores com mais mapas jogados', {'fontsize':25})
                                                                                                                                    
```

```python
ax = df_top10_mapas.plot(x = 'nome',
                                                                y = 'kd', 
                                                                kind ='line', 
                                                                color='green',
                                                                figsize =(18, 6),
                                                                fontsize=14)
ax.set_title('Os jogadores com os kd mais alto', {'fontsize':25})
```

```python
ax = df_top10_mapas.plot(x = 'nome',
                                                                y = 'kd diff', 
                                                                kind ='line', 
                                                                color='red',
                                                                figsize =(18, 6),
                                                                fontsize=14)
ax.set_title('Os jogadores com os kd diff mais alto', {'fontsize':25})
```

```python
ax = df_top10_mapas.sort_values(by='rating', ascending=False).plot(x = 'nome',
                                                                y = 'rating', 
                                                                kind ='barh', 
                                                                color='blue',
                                                                figsize =(18, 6),
                                                                fontsize=15)
ax.set_title('Os jogadores com os rating mais alto', {'fontsize':25})
```
